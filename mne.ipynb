```python
import mne
import scipy.signal as signal
import matplotlib.pyplot as plt

# Load the raw EDF file
raw = mne.io.read_raw_edf(r"D:\Clg\Project\SDP 2\dataset\16\physionet.org\files\hmc-sleep-staging\1.0.0\recordings\SN154.edf", preload=True)

# Pick the EEG C4-M1 channel
eeg_c4_m1 = raw.copy().pick_channels(['EEG C4-M1'])

# Get the raw EEG data and times from the selected channel
eeg_data, times = eeg_c4_m1[:]

print("\n\n")
print(eeg_data[0])
print("\n\n")

# Define the bandpass filter limits (0.5 - 4 Hz)
low, high = 0.5, 4

# Get the sampling frequency from the raw.info dictionary
fs = eeg_c4_m1.info['sfreq']  # Corrected key is 'sfreq'

# Design the Butterworth bandpass filter
b, a = signal.butter(3, [low, high], btype="bandpass", fs=fs)

# Apply the filter to the EEG data using filtfilt
filtered_signal = signal.filtfilt(b, a, eeg_data[0])

# Plotting the original and filtered EEG signals
plt.figure(figsize=(12, 6))

# Plot original EEG signal
plt.subplot(2, 1, 1)
plt.plot(times, eeg_data[0], label='Original EEG Signal (C4-M1)', color='blue')
plt.title('Original EEG Signal (C4-M1)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (µV)')
plt.grid(True)

# Plot filtered EEG signal
plt.subplot(2, 1, 2)
plt.plot(times, filtered_signal, label='Filtered EEG Signal (C4-M1)', color='green')
plt.title('Filtered EEG Signal (C4-M1)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (µV)')
plt.grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

plt.figure(figsize = (12,6))
plt.plot(times, delta_band.get_data()[0][:len(times)], label='Delta (0.5-4 Hz) Signal', alpha=0.7)
plt.show()

plt.figure(figsize = (12,6))
plt.plot(times, beta_band.get_data()[0][:len(times)], label='Beta (12-30 Hz) Signal', alpha=0.7)
plt.show()
```
